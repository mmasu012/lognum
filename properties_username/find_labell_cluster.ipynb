```python
labels = list(corpus['label'].unique())
for label in labels:
    print(label, corpus.loc[ corpus['label'] == label, 'name_struct'].shape)
    my_list = list(corpus.loc[ corpus['label'] == label, 'name_struct'].unique())
    with open('fb_' + str(label) + '_label.txt', 'w') as f:
        for item in my_list:
            f.write("%s\n" % item)
```
